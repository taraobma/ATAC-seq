## Results

### Sequencing Quality Control
Sequencing QC for the raw ATAC‑seq libraries showed high per‑base Phred scores across the full read length and consistent per‑sequence quality and GC content, with the main issue being clear adapter contamination and a drop in quality toward the 3′ ends. This was addressed by trimming adapters and low‑quality bases with Trimmomatic v0.39, which removed roughly 15–25% of reads per sample and substantially reduced overrepresented adapter sequences while preserving a tight GC range of ~47–49%. Post‑trimming MultiQC v1.19 reports indicated that all remaining FastQC v0.11.9 modules passed, duplication levels were moderate, and read length distributions were uniform at 50–51 bp, suggesting that the cleaned libraries were technically robust and suitable for downstream alignment and peak calling.

Alignment to the mm10 reference genome was highly successful, with only ~0.4–0.5% of reads unmapped across the eight samples, corresponding to >99.5% mapping efficiency and 19.6–28.5 million uniquely mapped reads per library. Duplicate rates were modest (approximately 15.5–19.3%), and removal of duplicate and mitochondrial reads left a large number of high‑quality nuclear fragments for each sample, providing strong signal‑to‑noise for peak calling and differential accessibility testing. These alignment statistics indicate that the libraries are well matched to the reference genome and that the downstream ATAC‑seq analyses (TSS enrichment, motif finding, and locus‑specific accessibility) can be interpreted with high confidence.

### QC Results and Alignment Statistics
Initial quality control for the ATAC-seq reads revealed generally high per-base sequence quality and consistent GC content, but flagged the presence of adapter sequences which led to a drop in quality scores. To address this issue, the raw reads were processed using Trimmomatic v0.39 to remove low-quality bases and residual adapter sequences prior to alignment. This trimming step was highly effective, with the percentage of dropped reads ranging from 14.8% to 24.9% across the samples, which successfully cleaned the data. The post-trimming MultiQC v1.19 summary confirmed the improvement, showing consistent quality profiles and a tight distribution of GC content between 47.0% and 49.0%, indicating that the libraries were of high technical quality and free from major contamination, ensuring the retained data was suitable for downstream analysis.

The alignment statistics demonstrate a highly successful mapping outcome, which provides a robust foundation for the downstream accessibility analysis. The key alignment metric showed an extremely low percentage of unmapped reads, ranging from only 0.4% to 0.5% across all eight samples, translating to a near-perfect alignment efficiency of over 99.5% to the mm10 reference genome. This confirmed that the trimmed libraries were highly compatible with the reference and that the vast majority of reads were biologically informative. The total number of uniquely mapped reads ranged from 19.6 million to 28.5 million, with moderate duplication rates (15.5% to 19.3%). By filtering out these duplicate and mitochondrial reads post-alignment, the remaining uniquely mapped reads provided a higher signal-to-noise ratio, which is essential for accurate peak calling and increases confidence in the reliability and biological interpretability of the identified accessible chromatin regions.




### ATAC-Seq QC Metrics
#### Fragment Size Distribution
The plots for every sample (cDC1 and cDC2, WT and KO) consistently display a uniform and singular, sharp peak at 50 bp. This pattern is not an accurate representation of the full DNA fragment lengths generated by the Tn5 transposase, but rather an artifact of the 50 bp read length used in the single-end sequencing protocol. Because single-end sequencing only captures one end of the DNA fragment, the bioinformatics pipeline cannot determine the actual distance between the two Tn5 insertion sites (the full fragment size). Consequently, the "fragment length" in these plots defaults to the maximum read length after trimming, resulting in the overwhelming peak at 50 bp. This severely limits the capacity for quality assessment: a high-quality ATAC-seq experiment relies on a multimodal distribution with periodic peaks at ~200 bp, ~400 bp, and beyond (reflecting mono- and di-nucleosomes) to confirm successful library preparation and high biological signal. The absence of these characteristic nucleosomal peaks is expected with single-end data, but means that the key nucleosome phasing quality metric, which is a standard measure of signal-to-noise ratio and library complexity cannot be reliably calculated or confirmed, thereby reducing confidence in the global quality validation of the accessible chromatin landscape.

#### Heatmap of the accessible regions
The TSS heat maps of the two celltypes demonstrate the expected hallmark of high‑quality ATAC‑seq data: a sharp, symmetric enrichment of signal centered precisely at transcription start sites, with signal gradually decaying into flanking regions. Within each cell type, the WT and KO replicates show nearly identical average profiles and very similar banding patterns in the heat maps, indicating good library complexity, consistent nucleosome‑free accessibility at promoters, and minimal technical variation between replicates. The broadly similar TSS profiles between WT and KO suggest that global promoter accessibility is largely preserved, implying that any regulatory effects of the knockout are likely to manifest as localized gains or losses of accessibility at specific subsets of genes rather than as a genome‑wide shift in chromatin openness.

#### Motifs Enrichment Results

Across all samples, the top motifs are ETS family binding sites, with very strong enrichment for PU.1 (SPI1), SpiB, ELF4/5, ETS1/3, and EHF motifs, indicating that ETS transcription factors are the dominant regulators at accessible regions in both cDC1 and cDC2. Many peaks also contain IRF8/PU.1 composite motifs and CTCF/CTCFL sites, suggesting cooperative regulation between ETS and IRF factors at immune regulatory elements and the presence of architectural CTCF sites within the accessible chromatin landscape. The high percentage of target sequences carrying these motifs, compared with background, supports a model in which classical dendritic‑cell lineage factors (PU.1, SpiB, IRF8, ELF proteins) and genome organizers (CTCF) collectively shape the differential chromatin accessibility observed between WT and KO conditions.

### Conclusion 


The overall reproduction of the panels from the original publication was not very successful. Specifically, the reproduction of the article's Figure 6C and 6E was not successfully completed.

My heatmap and the article's Figure 6A and 6B are visually different, but both support the same fundamental biological conclusion: the promoter regions remain strongly accessible in all dendritic-cell conditions, and the knockout does not globally disrupt transcription start site openness.

The key interpretive difference is that the article’s “gain” and “loss” panels directly point to distinct subsets of regions whose accessibility increases or decreases with HDAC1 loss. In contrast, my heatmap shows a robust, symmetric transcription start site enrichment in both wild-type and knockout samples without obvious global shifts of gain and loss. This implies that any knockout effects in my data are more likely confined to specific loci or regulatory modules rather than reflecting a genome-wide collapse or expansion of chromatin accessibility.

Finally, the Genome View of the MAGED1 gene was a success, as it clearly shows increased chromatin accessibility and gene expression of the MAGED1 gene in the knockout condition.
